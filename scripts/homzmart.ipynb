in the next step i convert volumn values for each column to list to now thrown any information as i can, and to be able to make unique row for each hotel(name), as the assessment didn't mention how to aggregate the data

```python
# copy the shape of the df with empty rows
ndf = df[df["categories"]!="Hotels"].copy()

# copy the unique names to the ndf (new_df), to be able to make one row and only one for each hotel
ndf.name = pd.Series(df.name.unique())

# for each row(hotel) make list of each column values
for col in df.columns.drop("name"):
    print(col)
    ndf[col] = [df[df.name==nam][col].unique() for nam in ndf.name]
ndf.shape    
```
